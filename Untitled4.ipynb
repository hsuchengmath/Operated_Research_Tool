```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP')
```

```python
infinity = solver.infinity()
# x and y are integer non-negative variables.
x = solver.IntVar(0.0, infinity, 'x')
y = solver.IntVar(0.0, infinity, 'y')
print('Number of variables =', solver.NumVariables())
```

```python
# x + 7 * y <= 17.5.
solver.Add(x + 7 * y <= 17.5)

# x <= 3.5.
solver.Add(x <= 3.5)

print('Number of constraints =', solver.NumConstraints())
```

```python
# Maximize x + 10 * y.
solver.Maximize(x + 10 * y)
```

```python
status = solver.Solve()
```

```python
if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    print('x =', x.solution_value())
    print('y =', y.solution_value())
else:
    print('The problem does not have an optimal solution.')
```

```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP')
infinity = solver.infinity()
```

```python
import random
#
user_list = [i for i in range(1000)]
con_list = [i for i in range(600)]
#
U2C2P = dict()
P_range = [0,1,0.5,0.6,0.7,0.8,0.9]
for u in user_list:
    U2C2P[u] = dict()
    for k in con_list:
        U2C2P[u][k] = random.sample(P_range ,1)[0]
#
salary_range = [10000,20000,15000]
salary_list = [random.sample(salary_range, 1)[0] for _ in range(len(con_list))]
#
session_num = 100
subR_list = []
for i in range(session_num):
    used_user = user_list[i*session_num : (i*2)+session_num]
    element = dict()
    for u in used_user:
        element[u] = None
    subR_list.append(element)
#
required_con_num = []
for _ in range(session_num):
    required_con_num.append(5)
```

```python
BH_subRindex2con = dict()
for i,subR in enumerate(subR_list):
    user_in_subR = list(subR.keys())
    BH_subRindex2con[i] = dict()
    for con in con_list:
        BH_subRindex2con[i][con] = 0
        for u in user_in_subR:
            BH_subRindex2con[i][con] += U2C2P[u][con]
```

```python
# build var
var_C_dict = dict()
for h in range(len(subR_list)):
    var_C_dict[h] = dict()
    for k in con_list:
        var_C_dict[h][k] = \
            solver.IntVar(0.0, infinity, "C_"+str(h)+'_'+str(k))
            #m.addVar(vtype=GRB.BINARY, name="C_"+str(h)+'_'+str(k))
```

```python
#operation = \
#    quicksum(BH_subRindex2con[h][k] * var_C_dict[h][k] for h in range(len(subR_list)) for k in con_list)
#m.setObjective(operation, GRB.MAXIMIZE)

objective = 0
for h in range(len(subR_list)):
    for k in con_list:
        objective += BH_subRindex2con[h][k] * var_C_dict[h][k]
solver.Maximize(objective)
```

```python
# build constrain
# first term
for k in con_list:
    first_constrain = 0
    for h in range(len(subR_list)):
        first_constrain += var_C_dict[h][k]
    solver.Add(first_constrain <= 1)
# second term
second_constrain = 0
for h in range(len(subR_list)):
    for k in range(len(con_list)):
        second_constrain += var_C_dict[h][k]*salary_list[k]
solver.Add(second_constrain <= 500 * 15000)
# third term
for h in range(len(subR_list)):
    third_constrain = 0
    for k in con_list:
        third_constrain += var_C_dict[h][k]
    solver.Add(third_constrain == required_con_num[h])
# forth term
for h in range(len(subR_list)):
    for k in con_list:
        solver.Add(var_C_dict[h][k] >=0)
for h in range(len(subR_list)):
    for k in con_list:
        solver.Add(var_C_dict[h][k] <= 1)      
```

```python
status = solver.Solve()
```

```python
if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    #print('x =', x.solution_value())
    #print('y =', y.solution_value())
else:
    print('The problem does not have an optimal solution.')
```
