```python
# build env
import random
user_list = [i for i in range(200)]
eng_lv_set = set([i+1 for i in range(12)])
attend_level_list = [random.sample(list(eng_lv_set), 1)[0] for _ in range(len(user_list))]
constrain_num_list = [i+2 for i in range(5)]
# init constrain_num2user_id_list
constrain_num2user_id_list = dict()
for constrain_num in constrain_num_list:
    constrain_num2user_id_list[constrain_num] = list()
for u in user_list:
    constrain_num2user_id_list[random.sample(constrain_num_list ,1)[0]].append(u)
```

```python
user2constrain_num = dict()
for constrain_num in constrain_num_list:
    ulist = constrain_num2user_id_list[constrain_num]
    for u in ulist:
        user2constrain_num[u] = constrain_num
```

```python
user2eng_lv = dict()
for i, u in enumerate(user_list):
    user2eng_lv[u] = attend_level_list[i]
```

```python
user2feasible_user = dict()
for u1 in user_list:
    user2feasible_user[u1] = list()
    for u2 in user_list:
        if u1 != u2 and abs(user2eng_lv[u1] - user2eng_lv[u2]) <= 1:
            user2feasible_user[u1].append(u2)
```

```python
from ortools.linear_solver import pywraplp
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP')
```

```python
# build var
var_U_dict = dict()
for u1 in user_list:
    var_U_dict[u1] = dict()
    for u2 in user_list:
        # diganal 
        if u1 == u2:
            variable = solver.IntVar(0.0, 0.0, "U_"+str(u1)+'_'+str(u2))
        else:
            # eng lv
            if abs(user2eng_lv[u1] - user2eng_lv[u2]) <= 1:
                variable = solver.IntVar(0.0, 1.0, "U_"+str(u1)+'_'+str(u2))
            else:
                variable = solver.IntVar(0.0, 0.0, "U_"+str(u1)+'_'+str(u2))            
        var_U_dict[u1][u2] = variable
```

```python
objective = 0
for u1 in user_list:
    for u2 in user_list:
        objective += var_U_dict[u1][u2]
solver.Maximize(objective)
```

```python
# build constrain
# first term
for i in range(len(user_list)):
    for j in range(len(user_list)):
        solver.Add(var_U_dict[i][j] - var_U_dict[j][i] == 0)
# second_term
for u1 in user_list:
    constrain_num = user2constrain_num[u1]
    second_term = 0
    for u2 in user_list:
        second_term += var_U_dict[u1][u2]
    solver.Add(second_term <= constrain_num) 
# second_term
for u1 in user_list:
    constrain_num = user2constrain_num[u1]
    second_term = 0
    for u2 in user_list:
        second_term += var_U_dict[u1][u2]
    solver.Add(second_term >= 1) 
```

```python
# diagnal is zero (v)
# eng lv =-1 (v)
# tanspoise (v)
# smaller than constrain num for each row
# user-user
```

```python
status = solver.Solve()
```

```python
if status == pywraplp.Solver.OPTIMAL:
    print('Solution:')
    print('Objective value =', solver.Objective().Value())
    #print('x =', x.solution_value())
    #print('y =', y.solution_value())
else:
    print('The problem does not have an optimal solution.')
```
